```python
import pandas as pd

df = pd.read_csv('dados.csv', delimiter=';')

print(df.head())
```

```python
import matplotlib.pyplot as plt

df_bh_jovens = df[(df['idade'] < 30) & (df['cidade'] == 'Belo Horizonte')]
total_jovens_bh = len(df_bh_jovens)
media_idade_jovens_bh = df_bh_jovens['idade'].mean()
media_salario_jovens_bh = df_bh_jovens['salario'].mean()
profissoes_bh_jovens = df_bh_jovens['profissao'].value_counts()

print("Dados brutos:")
print(df_bh_jovens.head())
print(f"\nNúmero total de jovens em Belo Horizonte: {total_jovens_bh}")
print(f"Idade média: {media_idade_jovens_bh:.2f} anos")
print(f"Salário médio: R$ {media_salario_jovens_bh:.2f}")
```

```python
plt.figure(figsize=(10, 6))
profissoes_bh_jovens.plot(kind='bar', color='skyblue')
plt.title('Distribuição de Profissões de Jovens em Belo Horizonte')
plt.xlabel('Profissão')
plt.ylabel('Número de Pessoas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('distribuicao_profissoes_bh_jovens.png')

df_bh_jovens.to_csv('jovens_bh_relatorio.csv', index=False, sep=';')
```
